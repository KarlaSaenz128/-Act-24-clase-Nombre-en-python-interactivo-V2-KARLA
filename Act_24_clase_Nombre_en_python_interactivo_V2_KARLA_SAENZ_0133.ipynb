```python
class Yate:
    def __init__(self, nombre, eslora, tripulacion, puerto_base):
        self.nombre = nombre
        self.eslora = eslora
        self.tripulacion = tripulacion
        self.puerto_base = puerto_base

# 1. Configuración inicial del Yate
print("--- Configuración de tu Yate ---")
nom = input("Nombre del yate: ")
esl = input("Eslora (metros): ")
tri = input("Número de tripulantes: ")
pue = input("Puerto base: ")

mi_yate = Yate(nom, esl, tri, pue)

# 2. Aquí es donde tú creas tus propias opciones
print("\n--- Configura tus opciones personalizadas ---")
opciones_usuario = {}

while True:
    accion = input("\nEscribe el nombre de una acción (o escribe 'listo' para terminar): ")
    if accion.lower() == 'listo':
        break
    resultado = input(f"¿Qué debe pasar cuando elijas '{accion}'?: ")
    opciones_usuario[accion] = resultado

# 3. Menú interactivo con tus opciones
def menu_personalizado():
    while True:
        print(f"\n--- Menú del Yate {mi_yate.nombre} ---")

        # Convertimos tus opciones en una lista numerada
        lista_acciones = list(opciones_usuario.keys())

        for i, accion in enumerate(lista_acciones, 1):
            print(f"{i}. {accion}")
        print(f"{len(lista_acciones) + 1}. Salir")

        try:
            seleccion = int(input("\nSelecciona un número: "))

            if seleccion == len(lista_acciones) + 1:
                print("¡Hasta luego, capitán!")
                print("elaborado por Karla Saenz 0133.")
                break

            # Ejecutamos la opción que tú escribiste
            accion_elegida = lista_acciones[seleccion - 1]
            print(f"\n>>> EJECUTANDO: {opciones_usuario[accion_elegida]}")

        except:
            print("elaborado por Karla Saenz 0133.")


menu_personalizado()
```
