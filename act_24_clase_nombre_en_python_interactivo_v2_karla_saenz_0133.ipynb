```python
class Yate:
    def __init__(self, nombre, eslora, tripulacion, puertoBase):
        # 'self' permite acceder a las propiedades únicas de este yate
        self.nombre = nombre
        self.eslora = eslora
        self.tripulacion = tripulacion
        self.puertoBase = puertoBase
        self.en_mar = False

    def navegar(self):
        if not self.en_mar:
            self.en_mar = True
            print(f"\n🌊 El '{self.nombre}' ha zarpado con éxito desde {self.puertoBase}.")
        else:
            print("\n⚠️ El yate ya se encuentra navegando.")

    def anclar(self):
        if self.en_mar:
            self.en_mar = False
            print(f"\n⚓ Ancla desplegada. El '{self.nombre}' se ha detenido.")
        else:
            print("\n⚠️ No puedes anclar, ya estás en el puerto.")

    def organizarEvento(self):
        print(f"\n--- Preparando fiesta en el '{self.nombre}' ---")
        # 'for' para recorrer los preparativos
        tareas = ["Limpieza de cubierta", "Catering de lujo", "Música en vivo"]
        for t in tareas:
            print(f"✅ Realizando: {t}...")
        print(f"🎉 ¡Evento listo para los {self.tripulacion} miembros de la tripulación y sus invitados!")

# --- Bloque Principal ---

def iniciar_sistema():
    print("Configuración inicial del Yate")
    n = input("Nombre del yate: ")
    e = input("Metros de eslora: ")
    t = int(input("Número de tripulantes: "))
    p = input("Puerto base: ")

    yate_usuario = Yate(n, e, t, p)

    # 'while True' para mantener el programa corriendo
    while True:
        print("\n--- MENÚ DE COMANDOS ---")
        print("1. navegar")
        print("2. anclar")
        print("3. evento")
        print("4. info")
        print("5. salir")

        accion = input("¿Qué acción desea realizar?: ").lower()

        # Uso de if y elif para decidir la función a ejecutar
        if accion == "1" or accion == "navegar":
            yate_usuario.navegar()
        elif accion == "2" or accion == "anclar":
            yate_usuario.anclar()
        elif accion == "3" or accion == "evento":
            yate_usuario.organizarEvento()
        elif accion == "4" or accion == "info":
            print(f"\nYate: {yate_usuario.nombre} | Puerto: {yate_usuario.puertoBase}")
        elif accion == "5" or accion == "salir":
            print("Cerrando sistema náutico. ¡Adiós!")
            print("Elaborado por Karla Saenz 0133")
            break
        else:
            print("❌ Comando no reconocido.")

if __name__ == "__main__":
    iniciar_sistema()
```
